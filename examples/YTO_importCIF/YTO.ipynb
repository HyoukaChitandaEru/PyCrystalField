# Import crystal structure of $\rm Yb_2Ti_2O_7$ from cif file and calculate the crystal field Hamiltonian

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

import PyCrystalField as cef
```

```python
########### Import CIF file

YTOLig, Yb = cef.importCIF('yto.cif','Yb1')
```

```python
########### print eigenvectors

Yb.diagonalize()
Yb.printEigenvectors()
```

```python
########### plot neutron spectrum

hw = np.linspace(0,100,200)
intens = Yb.neutronSpectrum(hw, Temp=20, Ei=120, ResFunc= lambda x: 4, gamma = 1)

plt.figure()
plt.plot(hw, intens)
plt.show()
```
