# Try fitting $\chi' = \frac{\chi}{1-\lambda \chi}$

```python
def chifitfun(x, lam):
    return x/(1-lam*x)

f, ax = plt.subplots(3,1, figsize=(6,7), sharex=True)

subplotlabels='abcdefg'
Temps = [NMTemp, NZTemp, PMTemp]
Chiminus1s = [NMChiminus1, NZChiminus1, PMChiminus1]
Calcsusceps = [NMcalcsuscep, NZcalcsuscep, PMcalcsuscep]
colors = [NMcol, NZcol, PMcol]
CompoundNames = ['$\\rm{Nd_3Sb_3Mg_2O_{14}}$', '$\\rm{Nd_3Sb_3Zn_2O_{14}}$', '$\\rm{Pr_3Sb_3Mg_2O_{14}}$']
yaxeslabels = ['$\chi^{-1}$ ($\mu_B$/T/Nd)$^{-1}$', '$\chi^{-1}$ ($\mu_B$/T/Nd)$^{-1}$', 
               '$\chi^{-1}$ ($\mu_B$/T/Pr)$^{-1}$']
Fields = ['0.5 T', '0.5 T', '0.1 T']

for i,axx in enumerate(ax):
    axx.plot(Temps[i],Chiminus1s[i], 's', markersize=4,mew=0.5,mec=colors[i]*np.array([0.5,0.5,0.5,1]), 
             c=colors[i], label='data')
    axx.plot(Temps[i],-1/chifitfun(Calcsusceps[i],2), label='calculation', c=fitcolor)
    axx.plot(Temps[i],-1/Calcsusceps[i], label='calculation', c='grey')
    axx.set_ylabel(yaxeslabels[i])
    axx.text(0.5,0.96,CompoundNames[i],
            horizontalalignment='center',verticalalignment='top', transform=axx.transAxes)
    
    axx.text(0.02,0.94, '('+subplotlabels[i]+')', fontsize=16,
        horizontalalignment='left',verticalalignment='top', transform=axx.transAxes)
    axx.text(0.98,0.6, Fields[i],
        horizontalalignment='right',verticalalignment='bottom', transform=axx.transAxes)
    axx.legend(loc=4, handlelength=1.5)

ax[2].set_xlabel('$T$ (K)')
ax[2].set_xticks([0,100,200,300])
    
plt.tight_layout()
plt.subplots_adjust(hspace=0.01)
```
