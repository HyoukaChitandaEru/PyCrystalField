```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import plotformat as pf

import sys
from pythonlib.functions import *
```

```python
ions = ['Sm3+','Pm3+','Nd3+','Ce3+','Dy3+','Ho3+','Tm3+','Pr3+','Er3+','Tb3+','Yb3+']
# ions = ['Yb3+']
nn = 3
compound = 'Pyrochlore'

nnn=[2,4,4,6,6,6]
mmm=[0,0,3,0,3,6]

for ion in ions:
    print('*'*40, ion)

    filename = ion[:2]+'TO'

    lowChisqP = np.load(compound+'/fitresults/{}{}_BestFit_uncertainty_P.npy'.format(filename,nn))
    lowChisqChi = np.load(compound+'/fitresults/{}{}_BestFit_uncertainty_Chi.npy'.format(filename,nn))
    lowChisqGtensor = np.load(compound+'/fitresults/{}{}_BestFit_uncertainty_Gtensor.npy'.format(filename,nn))
    lowChisqEvals = np.load(compound+'/fitresults/{}{}_BestFit_uncertainty_Evals.npy'.format(filename,nn))
    lowChisqEvecs = np.load(compound+'/fitresults/{}{}_BestFit_uncertainty_Evecs.npy'.format(filename,nn))

    bestfitB = lowChisqP[0][:6]


    ################# PLOT OUTPUT
    if ion in ['Sm3+', 'Ce3+']:
        f, ax = plt.subplots(2,2,figsize=(6.7,5.1), sharey=True)
        ax[1,1].axis('off')
    else:
        f, ax = plt.subplots(3,2,figsize=(6.7,8), sharey=True)

    markerprops = {'ls':'none', 'marker':'.', 'rasterized':True}

    scale = 1e3
    axes= ax.flatten()
    for i, axx in enumerate(axes):
        jind = i+3
        
        if (ion in ['Ho3+', 'Tm3+', 'Tb3+', 'Dy3+', 'Er3+']) and (i > 2):
            try:
                axx.plot(lowChisqP[:,jind]*scale*100, lowChisqChi, alpha=0.1, **markerprops)
                axx.plot(lowChisqP[0,jind]*scale*100, lowChisqChi[0], marker='.', ms=9, mfc='w', mew=1.3)
                axx.set_xlabel('$B_{}^{}$'.format(nnn[i],mmm[i])+' ($10^{-2} \\mu$eV)')
                axx.text(0.03,0.97,pf.parSubPlotLabel(i), va='top', transform=axx.transAxes)
            except IndexError: pass
        else:
            try:
                axx.plot(lowChisqP[:,jind]*scale, lowChisqChi, alpha=0.1, **markerprops)
                axx.plot(lowChisqP[0,jind]*scale, lowChisqChi[0], marker='.', ms=9, mfc='w', mew=1.3)
                axx.set_xlabel('$B_{}^{}$ ($\\mu$eV)'.format(nnn[i],mmm[i]))
                axx.text(0.03,0.97,pf.parSubPlotLabel(i), va='top', transform=axx.transAxes)
            except IndexError: pass
            
    ax[0,0].set_ylim(np.min(lowChisqChi)*0.9, np.max(lowChisqChi)*1.12)

    for i in range(len(ax)):
        ax[i,0].set_ylabel('$\\chi^2_{red}$')
        
    f.subplots_adjust(wspace=0.03, hspace=0.4, bottom=0.16, top=0.95)
    
    ax[0,1].text(0.95,0.97, '$\\rm'+ion[:2]+'_2Ti_2O_7$', ha='right', va='top', transform=ax[0,1].transAxes)
    SaveDirec = 'paperfigures/'+compound+'Comparisons/'
    plt.savefig(SaveDirec+ion[:2]+'2Ti2O7_chisquared.pdf', bbox_inches='tight', pad_inches = 0.02)
    plt.close()
```
